```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Input waarden met studierichtingen en aantal studenten
inputvalues = {
    'Industrial Engineering': [31],
    "Mechanical Engineering": [24],
    "Biomedical Engineering": [9],
    "Architecture urbanism and building sciences": [11],
    "Data Science": [6],
    "Electrical Engineering": [5],
    "MSc Human Technology Interaction": [2],
    "MSc Innovation Management": [3],
    "BSc Psychology & Technology": [3],
    "MSc Artificial Intelligence and Engineering": [2],
    "Bachelor Computer Science and Engineering": [4],
    "Bedrijfskunde": [1],
    "Mechanical Engineering Power and Flow": [1],
    "Operations Management & Logistics": [1],
    "Medical Engineering and Technology": [1],
    "Entrepreneurship and Business Innovation": [1],
    "Applied Physics": [3],
    "Other studie's": [46]
}

data = pd.DataFrame(inputvalues)

# Data herschikken zodat we categorieën en aantallen hebben
data = data.T  # Transponeren van de DataFrame zodat de studies in rijen komen
data.columns = ['Aantal Studenten']  # Kolom hernoemen

# Scheiden van "Other studie's"
other_studies = data.loc['Other studie\'s']
data = data.drop('Other studie\'s')  # Verwijder "Other studie's" uit de hoofddataset

# Sorteren van de data op het aantal studenten in aflopende volgorde
data = data.sort_values(by='Aantal Studenten', ascending=False)

# Voeg "Other studie's" weer toe aan het einde
data = data.append(other_studies)

# Maak de figuur en de achtergrondkleur instellen
fig, ax = plt.subplots()
fig.patch.set_facecolor('#afc1d0')  # Achtergrond kleur voor de figuur

# Breedte van de balken en de positie van de x-as instellen
bar_width = 0.85  # Smaller bar width for closer bars
indices = np.arange(len(data))  # Posities op de x-as

# Plotten van het staafdiagram met afgeronde hoeken
bars = ax.bar(indices, data['Aantal Studenten'], color='#1c3f60', width=bar_width)

# Afgeronde hoeken toepassen op de bars
for bar in bars:
    bar.set_linewidth(0)  # Geen randen
    bar.set_edgecolor('none')
    bar.set_capstyle('round')  # Voeg afgeronde hoeken toe aan de bars

# Tekstgrootte en rotatie van de x-as labels aanpassen
plt.xticks(indices, data.index, rotation=40, ha='right', fontsize=7)

# Labels en titel toevoegen
plt.xlabel('Studies', fontsize=12)
plt.ylabel('Aantal Studenten', fontsize=12)
plt.title('Aantal studenten per studie', fontsize=14)

# Pas de achtergrondkleur van de plot aan
ax.set_facecolor('#afc1d0')  # Achtergrond kleur voor de plot

# Zorg ervoor dat alles netjes wordt weergegeven
plt.tight_layout()

# Toon de plot
plt.show()


```
